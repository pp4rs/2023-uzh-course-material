```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
iris = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv")
iris.head()
```

```python
fig, ax = plt.subplots(2, 2)
sns.scatterplot(x='sepal_length', y='sepal_width', data=iris, ax=ax[1, 1])
fig.suptitle("A big title", fontsize=20)
```

```python
fig, ax = plt.subplots()
sns.scatterplot(x='sepal_length', y='sepal_width', hue='species',
                data=iris, ax=ax, alpha=0.5)
ax.set_xlabel("Sepal length")
```

```python
# fig, ax = plt.subplots()
sns.lmplot(x="sepal_width", y="sepal_length", hue="species", data=iris)
```

```python
sns.jointplot(x="sepal_width", y="sepal_length", data=iris, kind="kde")
```

```python
iris_long = iris.melt(id_vars="species")
```

```python
fig, ax = plt.subplots()
sns.violinplot(x="variable", y="value", hue="species", data=iris_long, ax=ax)
fig.set_size_inches(8, 5)
```

```python
sns.pairplot(iris, hue="species", kind="kde")
```
