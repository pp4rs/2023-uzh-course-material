```python
# generate data
import numpy as np

n = 10000
x = np.random.normal(size=n)
y = x**2 + np.random.normal(size=n)
```

```python
# plot data
import matplotlib.pyplot as plt

plt.scatter(x,y)
plt.savefig('first_plt_plot.png')
```

```python
# bit better way
fig, ax = plt.subplots()

ax.scatter(x,y)
ax.set_title('Lovely Plot')
ax.set_xlabel('x')
ax.set_ylabel('y')

plt.show()
```

```python
# bit better way
fig, ax = plt.subplots()

ax.scatter(x[x>0],y[x>0], color='red', marker='+')
ax.scatter(x[x<=0],y[x<=0], color='blue', marker='.')

ax.set_title('Lovely Plot')
ax.set_xlabel('x')
ax.set_ylabel('y')

plt.show()
```

```python
import statsmodels.formula.api as smf
import pandas as pd
data = pd.DataFrame({'x':x, 'y':y})
model = smf.ols(formula='y ~ x + I(x**2)', data=data)
fitted_model = model.fit()
fitted_model.summary()
```

```python
with plt.xkcd():
    y_hat = fitted_model.predict()

    order = np.argsort(x)

    fig, ax = plt.subplots()

    ax.scatter(x[x>0],y[x>0], color='red', marker='+')
    ax.scatter(x[x<=0],y[x<=0], color='blue', marker='.')
    ax.plot(x[order], y_hat[order], color='black')

    ax.set_title('Lovely Plot')
    ax.set_xlabel('x')
    ax.set_ylabel('y')

    plt.show()
```

```python
fig.set_size_inches(8,6)
fig.savefig('xkcd_plt.png', dpi=300)
```

```python
# bit better way
fig, ax = plt.subplots(nrows=1, ncols=2)

ax[0].scatter(x[x>0],y[x>0], color='red', marker='+')
ax[0].set_title('1')
ax[0].plot(x[order], y_hat[order], color='black')

ax[1].scatter(x[x<=0],y[x<=0], color='blue', marker='.')
ax[1].set_title('2')
ax[1].plot(x[order], y_hat[order], color='black')

fig.suptitle('Title')

plt.show()
```
