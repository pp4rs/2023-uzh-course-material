## MERGE

```python
dfs = [None, None, None, None]
for i in range(0,4):
    unemployment_rate_file_paths = glob.glob('../pandas-data/data/bls-employment-state/*0{}.csv'.format(i+3)) 
    list_of_files = [pd.read_csv(file) for file in unemployment_rate_file_paths]
    dfs[i] = pd.concat(list_of_files).reset_index(drop=True)
```

```python
unemployment_rates = dfs[0]
qty_unemployment = dfs[1]
qty_employment = dfs[2]
labor_force = dfs[3]
```

```python
unemployment_rates = dfs[0]
qty_unemployment = dfs[1]
qty_employment = dfs[2]
labor_force = dfs[3]

unemployment_rates = unemployment_rates.rename(columns={'value':'unemp_rate'}).drop('footnotes', axis='columns')
qty_unemployment = qty_unemployment.rename(columns={'value':'unemp_qty'}).drop('footnotes', axis='columns')
qty_employment = qty_employment.rename(columns={'value':'emp_qty'}).drop('footnotes', axis='columns')
labor_force = labor_force.rename(columns={'value':'labor_force'}).drop('footnotes', axis='columns')
```

```python
mergedData = unemployment_rates.merge(qty_unemployment, left_on=['state', 'year', 'period'], right_on=['state', 'year', 'period'], how='inner')
mergedData = mergedData.merge(qty_employment, left_on=['state', 'year', 'period'], right_on=['state', 'year', 'period'], how='inner')
mergedData = mergedData.merge(labor_force, left_on=['state', 'year', 'period'], right_on=['state', 'year', 'period'], how='inner')
```
